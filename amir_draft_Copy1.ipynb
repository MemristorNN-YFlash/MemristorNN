```python
import time
import numpy as np
from Activation import activeFunc
from Layer_amir import Layer
from Tests import sets
from Tests import Set
from prettytable import PrettyTable
        
class NeuralNetwork():
    def __init__(self,init_range,eta):
        self.init_range = init_range
        self.layers = []
        self.eta = eta
        self.errors = 0
        
    def add_layer(self,insize,outsize,actvfunc='Sigmoid'):
        self.layers.append(Layer(insize, outsize , activeFunc(actvfunc) , self.init_range))
    
    def forward_pass(self,inputdata):
        input_x = inputdata
        for x in self.layers:
            x.calc(input_x)
            input_x = x.out
    def back_propogation(self,x,y,batchsize,iter):
        L = len(self.layers)
        self.layers[-1].delta = (y - self.layers[-1].out)
        for l in reversed(range(1,L-1,1)):
            self.layers[l].delta_calc(self.layers[l-1].out ,self.layers[l+1],batchsize)
        self.layers[0].delta_calc(x,self.layers[1],batchsize)
    
    def epoch(self, epochs, batchsize, x_train, y_train, x_test, y_test):
        errvec = np.zeros([1,3])
        self.errors = np.zeros([epochs*batchsize,1])
        iter = 0
        for j in range(epochs):
            for i in range(batchsize):
                self.forward_pass(inputdata = x_train.current())
                errvec = self.layers[-1].out - y_train.current()
                self.errors[iter] = 0.5 * errvec @ errvec.transpose() 
                x_train.next_entry()
                self.back_propogation(x_train.current(),y_train.current(),batchsize,i)
                y_train.next_entry()
            #time.sleep(0.07)
                
            for x in self.layers: x.update_w(self.eta)
        
    def accuracy(self,x_test,y_test):
        correct_answers = 0
        for i in range(y_test.num_of_inputs):
            self.forward_pass(x_test.current())
            vec = self.layers[-1].out
            output1 = np.zeros_like(vec)
            output1[np.argmax(vec)]=1
            if (output1 == y_test.current()).all():
                correct_answers = correct_answers +1
            #print("NN output: ")
            #print(output1)
            #print("Target: ")
            #print(y_test.current())
            x_test.next_entry()
            y_test.next_entry()
        
        print(correct_answers/y_test.num_of_inputs)
    
    def summary(self):
        ## summary table of the NN
        i=0
        t = PrettyTable(['Layer number', 'Output shape', 'ActiveFunc', 'Weights shape'])
        for x in self.layers:
            t.add_row([i, x.outsize, x.actvfunc, x.w.shape])
            i = i+1
        print(t)
```

```python
x  = NeuralNetwork(0.5,0.1)
x.add_layer(4,3)
x.add_layer(3,3)
from sklearn import datasets
xx,yy = datasets.load_iris(return_X_y=True)
NN = sets(xx,yy,v1=0, v2=0.05)
x_train,x_test,y_train,y_test = NN.return_classes()
```

```python
x.accuracy(x_test, y_test)
x.epoch(100,10,x_train,y_train, x_test, y_test)
x.accuracy(x_test,y_test)
```
